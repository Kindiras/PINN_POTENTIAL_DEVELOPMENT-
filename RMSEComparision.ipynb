```python
dt7 = pd.read_csv("Erros7.csv",index_col=0)
dt9 = pd.read_csv("Erros9.csv",index_col=0)
dt11 = pd.read_csv("Erros11.csv",index_col=0)
dt112nd = pd.read_csv("Erros112nd.csv",index_col=0)
dt12 = pd.read_csv("Erros12.csv",index_col=0)
dt12
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(dt7["RMSError(eV/atom)"], 's-', color="black")
ax.plot(dt9["RMSError(eV/atom)"], 's-', color="purple")
ax.plot(dt11["RMSError(eV/atom)"], 's-', color="blue")
ax.plot(dt112nd["RMSError(eV/atom)"], 's-', color="red")
ax.plot(dt12["RMSError(eV/atom)"], 's-', color="green")
ax.set_xticks(range(len(dt11)))
ax.set_title("RMSE")
#ax.set_xticklabels(dt.index, rotation=45, ha='right')
ax.set_xticklabels(dt11.index, rotation=90)
ax.legend(["Oct-7","Oct-9","Oct-11","Oct-112nd","Oct-12"])
plt.tight_layout()
plt.xticks(fontsize=12)
plt.savefig("RMSE.pdf",dpi=6000)
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(dt7["Nconfigs"], 's-', color="black")
ax.plot(dt9["Nconfigs"], 's-', color="purple")
ax.plot(dt11["Nconfigs"], 's-', color="blue")
ax.plot(dt112nd["Nconfigs"], 's-', color="red")
ax.plot(dt12["Nconfigs"], 's-', color="green")
ax.set_xticks(range(len(dt11)))
ax.set_title("No. of Configurations")
#ax.set_xticklabels(dt.index, rotation=45, ha='right')
ax.set_xticklabels(dt11.index, rotation=90)
ax.legend(["Oct-7","Oct-9","Oct-11","Oct-112nd","Oct-12"])
plt.tight_layout()
plt.xticks(fontsize=12)
plt.savefig("Configuraiton.pdf",dpi=6000)
```
